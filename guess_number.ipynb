```python
### GUESS THE NUMBER

# Guess the number
import random


# function() : Assess user guess based on target and provide a hint
def check_guess(guess, target):
    
    if guess < target:
        return "The number is too low."
    elif guess > target:
        return "The number is too high"
    else:
        return "You've guessed the right number!"


# function() : Verify if user input is a number
def is_number(input_str):
    return input_str.isdigit()


# STAGE 1   SET THE GAME BEFORE PLAYING
while True:
    
    x = input("Choose the maximum number for guessing:")
    
    if is_number(x):
       x = int(x)
       print(f"The maximum number for guessing is {x}")
       break
    
    else:
        print("Please enter a valid number")

random_number = random.randint(1,x)


# STAGE 2   START AND KEEP PLAYING UNTIL CORRECT GUESS
while True:
    
    user_guess = input('Guess a number:')
    
    # verify whether user inputs a number ...
    if is_number(user_guess):
        guess = int(user_guess)
        
        # ....  and assess number 
        result = check_guess(guess, random_number)
        print(result)
        
        if result == "You've guessed the right number!":
            break
    
    # Catching loophole if user does NOT input a number (but a string or else instead)
    else:
        print("Please enter a valid number.")
        
```
